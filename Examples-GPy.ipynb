```python
def make_dict(key1,key2,value):
    from collections import defaultdict
    # Combine each of tuples into a single list
    l = zip(key1,key2,value)
    
    d = defaultdict(list)
    for key1,key2, value in l:
        d[(key1,key2)].append(value)
    return d
```
